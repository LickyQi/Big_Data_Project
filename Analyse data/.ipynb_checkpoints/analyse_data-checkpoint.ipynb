More detailed data description

```python
def description(df):
    summary = pd.DataFrame(df.dtypes,columns=['dtypes'])
    summary = summary.reset_index()
    summary['Name'] = summary['index']
    summary = summary[['Name','dtypes']]
    # Count the number of missing values in each column
    summary['Missing'] = df.isnull().sum().values    
    # Count the number of different values in each column, excluding null values
    summary['Uniques'] = df.nunique().values
    return summary
```

```python
des_train = description(X_train)
print(des_train)
```

```python
des_predict = description(X_predict)
print(des_predict)
```

Missing data & data type & data variance 

```python
# get the number of missing data points per column
missing_values_count = X_train.isnull().sum()

# look at the missing points in the first 15 columns
missing_values_count[0:15]
```

```python
# how many total missing values do we have?
total_cells = np.product(X_train.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
(total_missing/total_cells) * 100
```

```python
# get the number of missing data points per column
missing_values_count = X_predict.isnull().sum()

# look at the missing points in the first 15 columns
missing_values_count[0:15]
```

```python
# how many total missing values do we have?
total_cells = np.product(X_predict.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
(total_missing/total_cells) * 100
```

```python
print('Trainning set:')
print('- There are ',len(des_train['Missing'].unique())-1,' features with missing values;')
print('- There are ',(des_train['dtypes']=='object').sum(),' object-type features;')
print('- There are ',(des_train['dtypes']=='float64').sum(),' float-type features')
print('Name - Missing/Nb_row:')
des_train['Missing'] = des_train['Missing']/X_train.shape[0]
print(des_train[des_train['Missing']!=0].sort_values(by=['Missing'],ascending=False)[['Name','Missing']])
```

```python
print('Predict set:')
print('- There are ',len(des_predict['Missing'].unique())-1,' features with missing values')
print('- For each feature, there are ',des_predict['Missing'].unique()[1:],' values are missing;')
print('Name - Missing/Nb_row:')
des_predict['Missing'] = des_predict['Missing']/X_predict.shape[0]
print(des_predict[des_predict['Missing']!=0].sort_values(by=['Missing'],ascending=False)[['Name','Missing']])
```

for the feature with low missing values, we can use interpolation to fill them

for the features with high missing values (missing rate>0.4), perhaps we can delete them directly

Correlation between label and features

```python
X_temp = X_train
X_temp['Response'] = y_train
X_temp = X_temp.corr()
tabel_corr2 = abs(X_temp[['Response']]).sort_values(by=['Response'],ascending=False)
del X_temp
```
